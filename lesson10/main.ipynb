```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("jocker.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_background = cv2.imread("back.jpg")
img_background = cv2.cvtColor(img_background, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
```

```python
H, S, V = cv2.split(img_hsv)
```

```python
rows, cols = H.shape
img_background = cv2.resize(img_background, (cols, rows))

for i in range(rows):
    for j in range(cols):
        if 35 < H[i, j] < 85 and S[i, j] > 15 and V[i, j] > 80:
            img_rgb[i, j] = img_background[i, j]
```

```python
plt.imshow(img_rgb)
```
