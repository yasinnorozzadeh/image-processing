```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("im.png")
```

```python
rows, cols, _ = img.shape

for i in range(rows):
    for j in range(cols):
        B, G, R = img[i, j] 
        if B > 100 and G > 100 and R < 50:
            B = 0 
            G = 0
            R = 255
        if B > 100 and G < 50 and R > 100:
            B = 0 
            R = 0
            G = 255
        if B < 50 and G > 100 and R > 100:
            B = 255 
            G = 0
            R = 0
        img[i, j] = B, G, R
plt.imshow(img)
```
