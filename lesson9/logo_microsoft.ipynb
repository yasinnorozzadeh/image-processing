```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
logo = np.ones((600, 1000, 3), dtype = "uint8") * 127
plt.imshow(logo)
```

```python
logo[150:250, 150:250] = (246, 83, 20)

logo[260:360, 150:250] = (0, 161, 241)

logo[150:250, 260:360] = (124, 187, 0)

logo[260:360, 260:360] = (255, 187, 0)
```

```python
cv2.putText(logo, "Microsoft", (370, 300), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 255, 255), 8)
plt.imshow(logo)
```
