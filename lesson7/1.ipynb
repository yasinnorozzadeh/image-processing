```python
import cv2
import matplotlib.pyplot as plt
```

```python
# img_original = cv2.imread("0.jpg")
img_original = cv2.imread("0.jpg", cv2.IMREAD_GRAYSCALE)
_, img = cv2.threshold(img_original, 80 ,255, cv2.THRESH_BINARY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img = 255 - img
plt.imshow(img, cmap="gray")
```

```python
contours , _ =cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
```

```python
number_tas = 0
for c in contours:
    if len(c) > 100:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(img_original, (x, y), (x+w, y+h), (0, 255, 255), 3)
        number_tas += 1
plt.imshow(img_original, cmap="gray")
print(number_tas)
```
