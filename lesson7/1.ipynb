```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('anime-wolf.jpg', cv2.IMREAD_GRAYSCALE)
_ , img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img, cmap='gray')
```

```python
def Find_Contours(img, mode):
    rows, cols = img.shape 
    if mode=='RETR_LIST':
        point = []
        for i in range(rows):
            object_in_row = False
            for j in range(cols):
                if img[i, j]==255 and object_in_row==False:
                    object_in_row = True
                    point.append([i, j])
                elif img[i, j]==0 and object_in_row==True:
                    object_in_row = False
                    point.append([i, j])
    return point
```

```python
point= Find_Contours(img, 'RETR_LIST')
for h in point:
    if len(h) > 5:
        x, y, w, h = cv2.boundingRect(h)
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 127, 0), 3)
plt.imshow(img)
```
