```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("havich.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
R, G, B = cv2.split(img)
```

```python
# 1.
R_new = G
G_new = R
B_new = G
```

```python
img_result = cv2.merge((B_new, G_new, R_new))
img_result = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)

plt.imshow(img_result)
```

```python
# 2.
R_new = B
G_new = G
B_new = B
```

```python
img_result = cv2.merge((B_new, G_new, R_new))
img_result = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)

plt.imshow(img_result)
```
